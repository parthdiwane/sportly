```python
import pandas as pd 

df = pd.read_csv('/Users/parth/coding/python/sportly/stats/tennis_atp/singles/atp_matches_net.csv')
df
```

```python
df = df.drop(['winner_name','loser_name'],axis='columns')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
str_vals = ['tourney_name', 'surface','tourney_level','winner_hand','winner_ioc','loser_hand','loser_ioc','round']
label_encoder_variables = []
for variable_name in str_vals:
    var_name = "nlabel_" + variable_name
    globals()[var_name] = LabelEncoder()
    label_encoder_variables.append(globals()[var_name])
```

```python
for i in range(len(str_vals)):
    df[str_vals[i] + "_n"] = label_encoder_variables[i].fit_transform(df[str_vals[i]])
```

```python
for val in str_vals:
    df = df.drop([val],axis='columns')
df
```
